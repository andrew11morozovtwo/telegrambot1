## Краткая Документация Бота для Телеграмм-Канала "Безопасность всегда"

### Введение

Этот бот предназначен для автоматического комментирования постов в телеграмм-канале "Безопасность всегда" с помощью API OpenAI. Он помогает администраторам канала информировать подписчиков о правилах безопасности и предоставлять полезные советы.

### Функционал

- **Автоматическое комментирование**: Бот комментирует посты в канале, учитывая контекст и предыдущие комментарии.
- **Логирование данных**: Все комментарии и взаимодействия записываются в файл для дальнейшего анализа.
- **Обработка сообщений разных типов**: Бот не только отвечает на текстовые сообщения, но и обрабатывает сообщения с фото, видео, голосовыми сообщениями и опросами.

### Команды

| Команда | Описание |
|---------|----------|
| `/start` | Выводит приветственное сообщение и информацию о боте. |
| `/help` | Отображает инструкции по использованию бота. |

### Настройка

1. **Добавление бота в канал**: Добавьте бота в качестве администратора в телеграмм-канал и в чат обсуждения.
2. **Использование команд**: Используйте команды для настройки поведения бота.

### Технические детали

- **API OpenAI**: Используется для генерации комментариев.
- **Технологии**: Python, Telegram Bot API, OpenAI API.
- **Логирование**: Данные записываются в CSV-файл для анализа.
